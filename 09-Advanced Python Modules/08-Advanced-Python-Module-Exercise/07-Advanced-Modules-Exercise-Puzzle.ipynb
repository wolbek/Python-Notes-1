```python
import zipfile
import os
import re
```

```python
zip_obj=zipfile.ZipFile('unzip_me_for_instructions.zip','r')
```

```python
zip_obj.extractall('instruction_extracted')
```

```python
os.listdir()
```

```python
num_list=[]
for folder, sub_folders, files in os.walk('instruction_extracted'):
    print("Currently looking at folder: "+ folder)
    print('\n')
    print("THE SUBFOLDERS ARE: ")
    for sub_fold in sub_folders:
        print("\t Subfolder: "+sub_fold )
    
    print('\n')
    
    print("THE FILES ARE: ")
    for f in files:
        print("\t File: "+f)
        new_file = open(f'{folder}/{f}','r').read()
        number = re.search(r'\d{3}-\d{3}-\d{4}',new_file)
        if number is None:
            continue
        else:
            num_list.append(number.group())
        
        
    print('\n')
```
